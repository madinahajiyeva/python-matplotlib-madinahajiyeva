```python
# Write a function which will return N highest rated movies of a given year.
# It should accept dataframe, year and movie count N and return a new dataframe with top movies from imdb.
# It should also be possible to get movies by metascore, instead of imdb rating.
task[5] = 5
```

```python
def get_top_movies(dataframe, year, N=10, by='IMDB_Rating'):
    """
    Returns the top N highest-rated movies of a given year.
    
    Parameters:
    - dataframe (pd.DataFrame): The dataset to search within.
    - year (int): The release year of movies to filter by.
    - N (int): The number of top movies to return (default is 10).
    - by (str): The rating metric to sort by ('IMDB_Rating' or 'Meta_score').
    
    Returns:
    - pd.DataFrame: A dataframe with the top N movies from the given year.
    """
    # Validate the sorting column
    if by not in ['IMDB_Rating', 'Meta_score']:
        raise ValueError("Invalid sorting criterion. Choose 'IMDB_Rating' or 'Meta_score'.")
    
    # Filter by the specified year
    filtered_df = dataframe[dataframe['Released_Year'] == year]
    
    # Sort by the specified rating column and get the top N movies
    top_movies = filtered_df.sort_values(by=by, ascending=False).head(N)
    
    return top_movies
```

```python
def highest_rated_by_decade(dataframe, by='IMDB_Rating'):
    """
    Finds and prints the highest-rated movie of each decade.
    
    Parameters:
    - dataframe (pd.DataFrame): The dataset to search within.
    - by (str): The rating metric to sort by ('IMDB_Rating' or 'Meta_score').
    
    Returns:
    - pd.DataFrame: A new dataframe with the highest-rated movie of each decade.
    """
    
    dataframe['Decade'] = (dataframe['Released_Year'] // 10) * 10

    # Group by decade and get the top-rated movie for each
    top_movies_by_decade = dataframe.groupby('Decade').apply(
        lambda df: get_top_movies(df, df['Released_Year'].iloc[0], N=1, by=by)
    ).reset_index(drop=True)

    return top_movies_by_decade

top_movies_by_decade = highest_rated_by_decade(imdb, by='IMDB_Rating')
print("Highest-rated movie of each decade (by IMDb Rating):")
print(top_movies_by_decade)
```

```python
def top_successful_directors(dataframe, N=5):
    """
    Returns the top N commercially most successful directors based on total gross revenue.

    Parameters:
    - dataframe (pd.DataFrame): The dataset to search within.
    - N (int): The number of top directors to return (default is 5).
    
    Returns:
    - pd.DataFrame: A dataframe with the top N most successful directors.
    """

    dataframe['Gross'] = pd.to_numeric(dataframe['Gross'], errors='coerce')
    
    
    director_gross = dataframe.groupby('director')['Gross'].sum().sort_values(ascending=False)
    
   
    top_directors = director_gross.head(N).reset_index()
    top_directors.columns = ['Director', 'Total_Gross']

    return top_directors

top_5_directors = top_successful_directors(imdb, N=5)
print("Top 5 Commercially Most Successful Directors:")
print(top_5_directors)
```

```python
def most_frequent_top_actors(dataframe, min_imdb_rating=8.0, min_metascore=90):
    """
    Finds actors who starred most frequently in top-rated movies (IMDb rating >= 8 and Metascore >= 90).
    
    Parameters:
    - dataframe (pd.DataFrame): The dataset to search within.
    - min_imdb_rating (float): Minimum IMDb rating to qualify as a top movie.
    - min_metascore (int): Minimum Metascore to qualify as a top movie.
    
    Returns:
    - pd.DataFrame: A dataframe with actors and their appearance count in top movies, sorted in descending order.
    """
    top_movies = dataframe[(dataframe['IMDB_Rating'] >= min_imdb_rating) & 
                           (dataframe['Meta_score'] >= min_metascore)]
    
    actors = pd.concat([top_movies['Star1'], top_movies['Star2'], 
                        top_movies['Star3'], top_movies['Star4']])
    
    actor_counts = actors.value_counts().reset_index()
    actor_counts.columns = ['Actor', 'Appearances']
    
    return actor_counts

top_actors = most_frequent_top_actors(imdb)
print("Actors with Most Appearances in Top Movies:")
print(top_actors.head(10))

# The actor at the top of the list
most_frequent_actor = top_actors.iloc[0]
print("\nMost Frequent Top Actor:")
print(most_frequent_actor)
```

```python
def top_successful_actors(dataframe, N=5):
    """
    Returns the top N commercially most successful actors based on total gross revenue.
    
    Parameters:
    - dataframe (pd.DataFrame): The dataset to search within.
    - N (int): The number of top actors to return (default is 5).
    
    Returns:
    - pd.DataFrame: A dataframe with the top N most successful actors.
    """

    actor_gross_df = dataframe.melt(
        id_vars=['Gross'], value_vars=['Star1', 'Star2', 'Star3', 'Star4'],
        var_name='Star_Position', value_name='Actor'
    )
    
    actor_gross_df = actor_gross_df.dropna(subset=['Gross', 'Actor'])
    
    actor_total_gross = actor_gross_df.groupby('Actor')['Gross'].sum().sort_values(ascending=False)
    
    top_actors = actor_total_gross.head(N).reset_index()
    top_actors.columns = ['Actor', 'Total_Gross']

    return top_actors


top_5_actors = top_successful_actors(imdb, N=5)
print("Top 5 Commercially Most Successful Actors:")
print(top_5_actors)
```

```python
def most_common_genre_by_decade(dataframe):
    """
    Finds the most common genre for each decade.
    
    Parameters:
    - dataframe (pd.DataFrame): The dataset to analyze.
    
    Returns:
    - pd.DataFrame: A dataframe showing the most common genre in each decade.
    """
    dataframe['Decade'] = (dataframe['Released_Year'] // 10) * 10

    exploded_df = dataframe.assign(genre=dataframe['genre'].str.split(', ')).explode('genre')

    genre_counts = exploded_df.groupby(['Decade', 'genre']).size().reset_index(name='Count')

    most_common_genre = genre_counts.loc[genre_counts.groupby('Decade')['Count'].idxmax()]

    return most_common_genre

common_genres_by_decade = most_common_genre_by_decade(imdb)
print("Most Common Genre by Decade:")
print(common_genres_by_decade)

# Most common genres for the 1970s and 1980s
genre_1970s = common_genres_by_decade[common_genres_by_decade['Decade'] == 1970]
genre_1980s = common_genres_by_decade[common_genres_by_decade['Decade'] == 1980]
print("\nMost Common Genre of the 1970s:", genre_1970s['genre'].values[0])
print("Most Common Genre of the 1980s:", genre_1980s['genre'].values[0])
```

```python
def find_great_but_not_well_known_movies(dataframe, min_rating=8.0, max_votes=20000, min_metascore=None):
    """
    Finds movies that are highly rated but not well known.
    
    Parameters:
    - dataframe (pd.DataFrame): The dataset to search within.
    - min_rating (float): Minimum IMDb rating to qualify as a "great" movie.
    - max_votes (int): Maximum number of votes to consider a movie "not well known."
    - min_metascore (int or None): Minimum Metascore to include movies with high critical acclaim (optional).
    
    Returns:
    - pd.DataFrame: A dataframe of movies meeting the criteria.
    """
    filtered_df = dataframe[(dataframe['IMDB_Rating'] >= min_rating) & (dataframe['No_of_Votes'] <= max_votes)]
    
    if min_metascore is not None:
        filtered_df = filtered_df[filtered_df['Meta_score'] >= min_metascore]
    
    return filtered_df[['Series_Title', 'IMDB_Rating', 'No_of_Votes', 'Meta_score']]

great_movies = find_great_but_not_well_known_movies(imdb)

great_critically_acclaimed_movies = find_great_but_not_well_known_movies(imdb, min_metascore=90)

print("Great Movies (Rating >= 8, Votes <= 20,000):")
print(great_movies)

print("\nGreat Critically Acclaimed Movies (Rating >= 8, Votes <= 20,000, Metascore >= 90):")
print(great_critically_acclaimed_movies)
```

```python
def director_statistics(dataframe, director_name):
    """
    Returns a dictionary of statistics about a director.
    
    Parameters:
    - dataframe (pd.DataFrame): The dataset to analyze.
    - director_name (str): The name of the director to analyze.
    
    Returns:
    - dict: A dictionary containing various statistics about the director.
    """
    # Filter dataframe for the specified director
    director_df = dataframe[dataframe['director'] == director_name]

    if director_df.empty:
        return {"error": f"No movies found for director: {director_name}"}
    
    director_df['Runtime'] = director_df['Runtime'].astype(str).str.extract(r'(\d+)').astype(float)
    
    director_df['Gross'] = pd.to_numeric(director_df['Gross'], errors='coerce')
    
    movie_titles = director_df[['Series_Title', 'Released_Year']].to_string(index=False)
    
    best_movies = director_df[director_df['IMDB_Rating'] == director_df['IMDB_Rating'].max()]['Series_Title'].tolist()
    worst_movies = director_df[director_df['IMDB_Rating'] == director_df['IMDB_Rating'].min()]['Series_Title'].tolist()
    
    avg_imdb_rating = director_df['IMDB_Rating'].mean()
    
    avg_metascore = director_df['Meta_score'].mean()
    
    avg_runtime = director_df['Runtime'].mean()
    
    total_gross = director_df['Gross'].sum()

    genres = director_df['genre'].str.split(', ').explode().unique().tolist()
    
    director_stats = {
        "Movie Titles and Years": movie_titles,
        "Best Rated Movies": best_movies,
        "Worst Rated Movies": worst_movies,
        "Average IMDb Rating": avg_imdb_rating,
        "Average Metascore": avg_metascore,
        "Average Runtime (minutes)": avg_runtime,
        "Total Gross Revenue": total_gross,
        "Genres": genres
    }
    
    return director_stats

director_info = director_statistics(imdb, "Christopher Nolan")
print("Director Statistics:")
for key, value in director_info.items():
    print(f"{key}: {value}")
```
